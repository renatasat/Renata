1. Имеется список из 20 случайных чисел от 0 до 100. Напишите функцию, которая разбивает этот список на $n$ списков по $k$ элементов и включает все эти списки в новый список. Используйте $k$ в качестве параметра функции.

Алгоритм решения:
1) Используем функцию randint из модуля random для того, чтобы создать список из 20 случайных чисел от 0 до 100

Append() добавляет в конец списка элемент, переданный ему в качестве аргумента. Вызывается чрез оператор .(точка). В данном задание нужны числа от 0 до 100. Len()-встроенный метод для нахождения длины списка. Следовательно,в теле функции используем цикл for, в котором перебирается список random_lst,используя функцию range(). Начинается отбор со значения 0, проходит по всей длине нашего random_lst с помощью метода len(), разбивая по заданному значению k_elements
5) Потом заносим в пустой список (result), который объявили выше, с попомщью функции append() элементы списка с помощью записи [j:j+k_elements], которая означает, что будет генерироваться n-список от j до j+следующий n-список

````python
from random import randint 
def div_list(random_lst,k_elements):
    result = []
    for j in range(0,len(random_lst),k_elements):
        result.append(random_lst[j:j+k_elements]) 
        return result
random_number = []#пустой список
for j in range(20):# в цикле for заполняем список рандомными значениями
    random_number.append(randint(0,100))
n = int(input("Введите количество списков: "))
k = int(input("Введите количество элементов в каждом списке: "))
k_elements = 5
div_list = div_list(random_number,k_elements)
print(div_list)
````

2. Напишите функцию, вычисляющую факториал числа, передаваемого в качестве параметра.

Алгоритм решения:
1) Создаём функцию factorial(), так как в задание сказано написать функцию. В качестве аргумента передаём number
2) Внутри функции добавляю условие, где если переданное число в аргумент number равняется единице, то возвращаем его
3) Если условие не сработало, то возвращается аргумент number, который умножается на вызов функции factorial() с передачей аргумента number-1
4) В конце вводим,вызываем функцию и выводим результат

````python
def factorial (number):
    if number == 1:
        return number
    else:
        return number * factorial(number-1)
print (factorial(int(input('Введите число:'))))
````

3. Написать функцию умножения, принимающую от одного до трех параметров. Функцию вызвать с приведенными ниже аргументами. Для случая a4 выбрать 3 первых и 3 последних значения.
```
a1 = (15, 10 ,5)
a2 = (3, 1)
a3 = [2, 35, 55]
a4 = (5, 10 ,15, 20)
```
Алгоритм решения:
1) Создаём функцию mul с использованием специального синтаксиса *args(позволяет функции принять переменное количество аргументов)
2) Переменная result(используется для хранения результата умножения аргументов) и присваиваем ей начальное число 1
3) В цикле for, который будет повторяться по всем аргументам, переданным в функцию,где каждый аргумент будет храниться в переменной num
4) Внутри цикла умножаем значения переменной(result) на текущей аргумент(num) и присваиваем результат обратно в переменную (result)
5) Результат умножения сохраняется в соответсвующих переменных result1, result2, result3 и result4, затем выводим результаты на экран
6) В случае a4, используются срезы a4[:3] для выбора первых трех элементов и a4[-3:] для выбора последних трёх элементов

````python
def mul(*args):
    result = 1
    for num in args:
        result *= num
    return result
a1 = (15,10,5)
a2 = (3,1)
a3 = [2,35,55]
a4 = (5,10,15,20)
result1 = mul(*a1)
print(result1)
result2 = mul(*a2)
print(result2)
result3 = mul(*a3)
print(result3)
result4 = mul(*a4[:3],*a4[-3:])
print(result4)
````

5. Реализовать функции для выполнения четырех арифметических операций, преобразующих целые числа в целое число. Создать словарь с функциями исоответствующими им симв олами операций. Для двух заранее заданных целых чисел (например, 25 и 4) выполнить выбранную пользователем арифметическую операцию.

Алгоритм решения:
1) Определяем 4 функции add, subtract, mul и divide для выполнения сложения, вычитания, умножения и деления
2) в функции divide() задаём условия, что если y равен 0, то деление невозможно, а если всё хорошо, то спокойно делим
3) Затем создаем словарь operations, где ключи-это символы (+,-,*,/), а значения-это функции, связанные с этими символами
4) Вводим оператор и сохраняем его в переменную operator 
5) Если оператор является ключом в словаре operations, то вызывается соответствующая функция из словаря,передавая значения number1 и number2(они даны нам уже в задание) в качестве аргумента

````python
def add(x,y):
    return x + y
def subtract(x,y):
    return x - y
def mul(x,y):
    return x * y
def divide(x,y):
    if y != 0:
        return x / y
    else:
        print("Деление на ноль невозможно")
operations = {
    '+': add,
    '-': subtract,
    '*': mul,
    '/': divide
}
number1 = 25
number2 = 4
operator = input("Введите опретор(+,-,*,/):")
if operator in operations:
    result = operations[operator](number1,number2)
    print("Результат: ",result)
````

6. Написать калькулятор для строковых выражений вида '<число> <операция> <число>', где <число> - целое число, например 113, <арифметическая операция> - одна из операций `+,-,*,//,%,^`. Пример: `calc('13 - 5') -> 8`

Алгоритм решения:
1) Создаём функцию calc(), где в качестве аргумента является exp
2) eval()-это встроенная функция, которая анализирует аргумент выражения и оценивает его как выражение.Следовательно, внутри функции используем функцию eval(), которая вычисляет строковое выражение. Затем возвращается результат вычисления выражения
3) None-значение, обозначающее нейтральное поведение. Try-попытаться, а Except-исключение (описание работы: попытаться сделать то-то и то-то, если при этом возникло исключение, то сделать вот это и это)
4) Ввод выражения и сохранение его в переменную exp
5) Вызываем функцию calc, передавая ей введенное выражение, сохраняем результат в переменую result

````python
def calc(exp):
    try:
        result = eval(exp)
        return result
    except:
        print ("Ошибка в выражении")
        return None
exp = input("Введите выражение: ")
result = calc(exp)
if result is not None:
    print ("Результат: ", result)
````

Алгоритм  решения:
1) Создаём словарь to_int(), где в качестве аргумента является text, который свяжет числительные на русском языке с их числовыми значениями

Split()-функция, которая используется для разделения строки на список строк на основе разделителя. Следовательно, words = text.lower().split(): Входная строка text приводится к нижнему регистру с помощью метода lower(), затем разделяется на отдельные слова с помощью метода split(). 
3) Начинаем цикл, где каждое слово из списка words поочередно присваивается переменной word
4)  if word in numbers: Происходит проверка, присутствует ли текущее слово word в словаре numbers. numbers - это словарь, где ключи являются словами, представленными в виде текста, а значения - числами
5)  number = to_int(number_text): Вызывается функция to_int с аргументом number_text, чтобы преобразовать введенный текст в числовое значение

````python
def to_int(text):
    numbers = {
         'ноль' :0, 'один' :1, 'два' :2, 'три' :3, 'четыре' :4, 'пять' :5, 'шесть' :6, 'семь' :7, 'восемь' :8, 'девять' :9, 'десять' :10,
        'одиннадацать' :11, 'двенадцать' :12, 'тринадцать' :13, 'четырнадцать' :14, 'пятнадцать' :15, 'шестнадцать' :16, 'семнадцать' :17, 'восемнадцать' :18, 'девятнадцать' :19,
        'двадцать' :20, 'тридцать' :30, 'сорок' :40, 'пятьдесят' :50, 'шестьдесят' :60, 'семьдесят' :70, 'восемьдесят' :80, 'девяносто' :90,
        'сто' :100, 'двести' :200, 'триста' :300, 'четыреста' :400, 'пятьсот' :500, 'шестьсот' :600, 'семьсот' :700, 'восемьсот' :800, 'девятьсот' :900
    }
    words = text.lower().split()  
    result = 0

    for word in words:
        if word in numbers:
            result += numbers[word]#Если текущее слово word присутствует в словаре numbers, то соответствующее ему значение прибавляется к переменной result.

    return result

number_text = input('Введите число в виде текста: ')
number = to_int(number_text)
print("Цифры: ",number)
````

И сообщает об ошибках (выводит на экран описание типа ошибки и возвращает число -1). Пример: to_int('тридцать три') -> 33 Пример: to_int("сто сорок тридцать два сто") -> -1 Вывод: тридцать - некорректное расположение в числе.

Алгоритм решения:
1) Словарь num содержит соответствие между текстовыми представлениями чисел и их числовыми значениями
2) Строка words = text.split() разбивает входной текст на слова и сохраняет их в список words.
3) Переменные number и total инициализируются нулевыми значениями и будут использоваться для накопления числа и общей суммы чисел соответственно
4) Переменная prev_num инициализируется значением None. Она будет использоваться для хранения предыдущего числа и проверки корректности записи чисел с использованием условия if r_num is not None and prev_num is not None and r_num > prev_num
5) Внутри цикла for каждое слово проверяется в словаре num с помощью num[word]. Полученное числовое значение сохраняется в переменную r_num. Если r_num не равно None, то это слово корректно представляет число. Далее, проверяем, чтобы текущее число r_num было не больше предыдущего числа prev_num, используя условие if r_num is not None and prev_num is not None and r_num > prev_num. Если текущее число больше предыдущего, возвращается сообщение о некорректной записи числа
6) Далее код проверяет, является ли текущее число r_num меньше 10. Если это так, проверяется, равно ли number нулю. Если да, то присваивается текущее число r_num к number, иначе число r_num добавляется к number после умножения number на 1 
7) В конце итерации текущее число number добавляется к общей сумме total
8) После завершения цикла возвращается общая сумма чисел total
9) Выводим результат на экран с помощью функции print()

````python
def to_int(text):
    num = {
        'ноль': 0, 'один': 1, 'два': 2, 'три': 3, 'четыре': 4,
        'пять': 5, 'шесть': 6, 'семь': 7, 'восемь': 8, 'девять': 9,
        'десять': 10, 'одиннадцать': 11, 'двенадцать': 12, 'тринадцать': 13, 'четырнадцать': 14,
        'пятнадцать': 15, 'шестнадцать': 16, 'семнадцать': 17, 'восемнадцать': 18, 'девятнадцать': 19,
        'двадцать': 20, 'тридцать': 30, 'сорок': 40, 'пятьдесят': 50,
        'шестьдесят': 60, 'семьдесят': 70, 'восемьдесят': 80, 'девяносто': 90,
        'сто': 100, 'двести': 200, 'триста': 300, 'четыреста': 400, 'пятьсот': 500,
        'шестьсот': 600, 'семьсот': 700, 'восемьсот': 800, 'девятьсот': 900
    }

    words = text.split()#разбиваем текст на слова
    #инициализируем переменные для хранения числа и суммы
    number = 0
    total = 0
    prev_num = None

    for word in words:
        r_num = num[word]
        if r_num is not None and prev_num is not None and r_num > prev_num:
            return f"{r_num} - некорректная запись числа"
        prev_num = num[word]


        if r_num < 10:#проверка,если число меньше десяти
            if number == 0:#если нет предыдущего числа,то добавим текущее
                number += r_num#если есть предыдущее число,то добавим к нему текущее
            else:
                number *= 1#умножаем на 1,чтобы получить правильное значение
                number += r_num

        else:
            if number != 0:#если есть число меньше 10,то лобавим его к общему числу
                total += number
            number = r_num
            prev_num = r_num
    #добавляем оставшееся число к сумме
    total += number
    return total


text = input("Введите текст: ")
result = to_int(text)
print(f"Результат: {result}")
````
