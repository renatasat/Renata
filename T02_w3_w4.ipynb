14. Реализовать калькулятор, который принимает от пользователя через пробел строку следующего вида: «a op b», где a и b – некоторые числа, а ‘op’ определяет оператор и может принимать значения $«+,-,*,/,**,%»$. В зависимости от оператора с помощью форматирования строк вывести результат в виде: «a + b = 3 + 2 = 5», где была получена строка «3 + 2». Для операций возведения в степень и деления по модулю использовать вместо знака оператора соответствующие выражения. Например, для строки «2 ** 3» должно быть выведено «a в степени b = 2 в степени 3 = 8».

````python
operation = input("Ввести операцию: ")
operands = operation.split(" ")#разделяет строку операции на отдлеьные элементы с попощью метода split()
a = float(operands[0])#извлекают первый и третий элементы из списка
b = float(operands[2])
operator = operands[1]#список,а [1] используется для доступа ко второму элементу  
if operator == "+":
    result = a + b
elif operator == "-":
    result = a - b
elif operator == "*":
    result = a * b
elif operator == "/":
    result = a / b
elif operator == "**":
    result = a ** b
elif operator == ".":
    result = a % b
else:
    print("Неподдерживаемый оператор")
output = f"{int(a)}{operator}{int(b)} = {int(a)}{operator}{int(b)} = {result}"#строка,включающая значения пременных в форматированном виде
print(output)
````

16. Имеются два n – мерных вектора x и y, которые задают координаты n точек на плоскости (случайные целые числа). Найти наиболее близкие друг другу точки.

````python
import random
n = 3
num_points = 5

x = []#создание пустого списка для хранения элементов вектора
for _ in range(num_points):#запускается цикл.
    point = [random.randint(1, 100) for _ in range(n)]#На каждой итерации функция random.randint(0,100) генерирует случайное число
    x.append(point)#Случайное число добавляется в конце списка vector с попомщью метода append

y = []
for _ in range(num_points):
    point = [random.randint(1, 100) for _ in range(n)]
    y.append(point)
def distance(point1, point2):
    squared_distance = sum((x - y) ** 2 for x, y in zip(point1, point2))
    return squared_distance ** 0.5

min_distance = None#инициализация переменной(присваивание начального значения,но неизвестно какое оно будет)
closest_points = None

for i in range(num_points):#Внешний цикл for i in range(num_points) перебирает числа от 0 до num_points - 1 и присваивает каждое значение переменной i.
    for j in range(num_points):#Внутренний цикл for j in range(num_points) похож на внешний цикл и также перебирает числа от 0 до num_points - 1, присваивая значение j.
        d = distance(x[i], y[j])
if min_distance is None or d < min_distance:#Выражение означает, что если min_distance равно None или d меньше min_distance, то выполняется условие
    min_distance = d
closest_points = (x[i], y[j])
print(f"Наиболее близкие точки: {closest_points[0]} и {closest_points[1]}")
print(f"Расстояние между ними: {min_distance:.2f}")
````

````python
N = 10
numbers = set(range(1, N + 1))#Создаём множества целых чисел от 1 до N
multiples_of_2 = {x for x in numbers if x % 2 == 0}#множество чисел, кратных 2
multiples_of_3 = {x for x in numbers if x % 3 == 0}
multiples_of_6 = {x for x in numbers if x % 2 ==0  and x % 3 == 0}
multiples_of_2_or_3 = {x for x in numbers if x % 2 == 0 or x % 3 == 0}
print("Множество чисел кратных 2:", multiples_of_2)
print("Множество чисел кратных 3:", multiples_of_3)
print("Множество чисел кратных 6:", multiples_of_6)
print("Множество чисел кратных 2 или 3:", multiples_of_2_or_3)
````

````python
words = ["mango", "strawberry"]#список слов
target_letter = "a"#в переменной задаём букву,которую хотим найти
count = 0#счётчик
for word in words:#перебераем каждое слово в списке
    for letter in word:#перебераем каждую букву в слове
        if letter == target_letter:#сравниваем букву с заданной буквой
            count += 1#если буквы совпадают, увеличиваем счётчик
print(f"Общее количество букв:{target_letter} в словах:{count}")
````

````python
words = ["mango", "strawberry"]#список слов
target_letter = "a"#в переменной задаём букву,которую хотим найти
max_percentage = 0#инициализируем переменные для хранения максимальной доли
word_with_max_percentage = ""#и соотвествующего слова
for word in words:#перебираем каждое слово в списке
    percentage = (word.count(target_letter) / len(word)) * 100#считаем долю заданной буквы в слове; word.count(target_letter) подсчёт количества раз буквы,встречающийся в строке
    if percentage > max_percentage:
        max_percentage = percentage
        word_with_max_percentage = word
print(f"Слово с наибольшей долей буквы:{target_letter} : {word_with_max_percentage}, {max_percentage:.2f}%")
#выводим слово с наибольшей долей заданной буквы и эту долю
````
