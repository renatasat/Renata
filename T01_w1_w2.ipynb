1.10 Имеются два $n$–мерных вектора $x$ и $y$, которые задают координаты $n$ точек на плоскости (случайные целые числа). Найти наиболее близкие друг другу точки.

````python
import random
n = 3
num_points = 5

x = []#создание пустого списка для хранения элементов вектора
for _ in range(num_points):#запускается цикл.
    point = [random.randint(1, 100) for _ in range(n)]#На каждой итерации функция random.randint(0,100) генерирует случайное число
    x.append(point)#Случайное число добавляется в конце списка vector с попомщью метода append

y = []
for _ in range(num_points):
    point = [random.randint(1, 100) for _ in range(n)]
    y.append(point)
def distance(point1, point2):
    squared_distance = sum((x - y) ** 2 for x, y in zip(point1, point2))
    return squared_distance ** 0.5

min_distance = None#инициализация переменной(присваивание начального значения,но неизвестно какое оно будет)
closest_points = None

for i in range(num_points):#Внешний цикл for i in range(num_points) перебирает числа от 0 до num_points - 1 и присваивает каждое значение переменной i.
    for j in range(num_points):#Внутренний цикл for j in range(num_points) похож на внешний цикл и также перебирает числа от 0 до num_points - 1, присваивая значение j.
        d = distance(x[i], y[j])
if min_distance is None or d < min_distance:#Выражение означает, что если min_distance равно None или d меньше min_distance, то выполняется условие
    min_distance = d
closest_points = (x[i], y[j])
print(f"Наиболее близкие точки: {closest_points[0]} и {closest_points[1]}")
print(f"Расстояние между ними: {min_distance:.2f}")
````

1.12 Пусть пользователь вводит 5 чисел: а, b, c, d, e. Реализуйте программу расчета выражения вида $res=\frac{a+bc}{2d-e}+Ost[b^e/c]$, где $Ost$ - остаток от деления. Учесть невозможность деления на 0.

````python
import sys
a = int(input("Ввести значения a: "))
b = int(input("Ввести значения b: "))
c = int(input("Ввести значения c: "))
if c == 0:
    sys.exit("Некорректное значение переменной c. Деление на ноль")#выкидывает из программы
d = int(input("Ввести значения d: "))
e = int(input("Ввести значения e: "))
part1 = a+b+c
part2 = 2*d-e
if part2 == 0:
    sys.exit("Некорректное значения. Деление на ноль при вычислении")
res = (part1/part2) + ((b**e)%c)
print("Результат: ", res)
````
